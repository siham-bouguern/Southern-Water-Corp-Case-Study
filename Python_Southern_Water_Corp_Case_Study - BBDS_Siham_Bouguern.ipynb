## Step 7: <span style="color:green">Identify Outliers</span> 

**Step 07: Identifying outliers**

**i. Calculating how many entries currently in the original dataframe**

```python
# creating a copy of the original dataframe
complete_dataframe_1 = pd.concat([dataframe_1, dataframe_2, dataframe_3])

# dropping the 'TIMEFRAME' attribute if it exists
if 'TIMEFRAME' in complete_dataframe.columns:
    complete_dataframe.drop(columns='TIMEFRAME', inplace=True)

# calculating the lower and upper bounds for outliers using the interquartile range IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# identify the outliers for each column
outliers = ((complete_dataframe < lower_bound) | (complete_dataframe > upper_bound)).sum()


# calculating the total number of entries in the original Dataframe
total_entries = complete_dataframe_1.shape[0]
print('The number of total entries is: ', total_entries)
```

**ii. Identifying the number of entries if the outliers were to remove**

```python
# calculating the number of entries without outliers
entries_without_outliers = total_entries - outliers
print('The number of entries without outliers is:\n', entries_without_outliers)

# calculating the number of outliers
num_outliers = outliers.sum()
print('The number of outliers is: ', num_outliers)
```

**iii. Finding the proportion of outliers that exist in the dataset**

```python
# calculating the proportion of outliers in the dataset
proportion_outliers = num_outliers / total_entries
print("Proportion of outliers in the dataset:", proportion_outliers)
```

# Part III: <span style="color:blue">Creating a Multivariate Regression Model</span>

## Step 14: <span style="color:purple">Use OLS Regression</span> 
**Open-ended Question:**

Which linear regression model seems to be a better fit?

**Step 14: Creating a Multivariate Regression Model**

**i. Creating a regression equation using OLS Regression**

```python
#Answer for step i): 
#Below is the first part of the code
combine = [dataframe_1, dataframe_2, dataframe_3]
dataframe_two = pd.concat(combine)
dependent_var = dataframe_two['PUMP FAILURE (1 or 0)']
filt = ['SURJEK_FLOW_METER_1', 'SURJEK_FLOW_METER_2', 'ROTATIONAL_PUMP_RPM',
       'SURJEK_PUMP_TORQUE', 'MAXIMUM_DAILY_PUMP_TORQUE',
       'SURJEK_AMMONIA_FLOW_RATE', 'SURJEK_TUBE_PRESSURE',
       'SURJEK_ESTIMATED_EFFICIENCY']
#----write your code below------

# Extract the independent variables excluding 'PUMP FAILURE (1 or 0)'
independent_vars = dataframe_two[filt]

# Drop rows with missing or infinite values in both the dependent and independent variables
dataframe_merged = pd.concat([dependent_var, independent_vars], axis=1)
dataframe_merged = dataframe_merged.dropna()

# Extract the dependent variable and independent variables after dropping missing values
dependent_var = dataframe_merged['PUMP FAILURE (1 or 0)']
independent_vars = dataframe_merged.drop(columns=['PUMP FAILURE (1 or 0)'])

# Add a constant term to the independent variables for the intercept term in the regression model
independent_vars = sm.add_constant(independent_vars)

# Fit the OLS model
model_1 = sm.OLS(dependent_var, independent_vars).fit()

# Get the R-squared value for the model
r_squared_1 = model_1.rsquared
print("R-squared for the model with all independent variables:", r_squared_1)

# Print the summary of the model
print()
print(model_1.summary())
```

**ii. Repeating step (i) using the rolling standard deviation**

```python
#Answer for step ii):

# creating the dataframe
combine = [dataframe_1, dataframe_2, dataframe_3]
dataframe_two = pd.concat(combine)

# identifying the dependent variable
dependent_var = dataframe_two['PUMP FAILURE (1 or 0)']

# creating a filter list
filt = ['SURJEK_FLOW_METER_1', 'SURJEK_FLOW_METER_2', 'ROTATIONAL_PUMP_RPM',
       'SURJEK_PUMP_TORQUE', 'MAXIMUM_DAILY_PUMP_TORQUE',
       'SURJEK_AMMONIA_FLOW_RATE', 'SURJEK_TUBE_PRESSURE',
       'SURJEK_ESTIMATED_EFFICIENCY', 'PUMP FAILURE (1 or 0)']

# create dataframes for the rolling standard deviation variables (created previously)
indep_roll_std30 = independent_vars.rolling(window=30).std()
indep_roll_std30.dropna(inplace=True)
indep_roll_std30 = indep_roll_std30.join(dataframe_merged[['PUMP FAILURE (1 or 0)']], how='inner')
indep_nonan = indep_roll_std30.loc[:, filt]
dep_nonan = indep_roll_std30.loc[:, ['PUMP FAILURE (1 or 0)']]

# creating regression model for the rolling standard deviation variables.
model_2 = sm.OLS(dep_nonan, indep_nonan).fit()

print(model_2.summary())
```
